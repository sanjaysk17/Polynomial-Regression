```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\sanja\Desktop\Polynomial Regression Task\bike_poly.csv")
df.columns
```

```python
x=df[['temperature']]
y=df['rentals']
```

```python
plt.scatter(df['temperature'],y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=26)
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
plt.scatter(x_train,y_train)
plt.plot(x_train,model.predict(x_train),color='red')
plt.title("The best Fit line In a Training Data")
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
score=r2_score(y_test,model.predict(x_test))
print(score)
print(root_mean_squared_error(y_test,model.predict(x_test)))
```
